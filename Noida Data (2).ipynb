```python
# Importing libraries
import requests
import pandas as pd
```

```python
header={"user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/112.0.0.0 Safari/537.36"}
url="https://vegetablemarketprice.com/api/data/market/noida/chartdata?start=2022-01-26&end=2023-05-10&vegIds=onionBig,onionSmall,tomato,greenChili,beetroot,potato,plantain,amaranthLeaves,amla,ashGourd,babyCorn,bananaFlower,capsicum,bitterGourd,bottleGourd,butterBeans,broadBeans,cabbage,carrot,cauliflower,clusterBeans,coconut,colocasiaLeaves,colocasia,corianderLeaves,corn,cucumber,curryLeaves,dillLeaves,drumsticks,brinjal,brinjalBig,elephantYam,fenugreekLeaves,frenchBeans,garlic,ginger,onionGreen,greenPeas,ivyGourd,lemon,mangoRaw,mintLeaves,mushroom,mustardLeaves,ladiesFinger,pumpkin,radish,ridgeGourd,shallot,snakeGourd,sorrelLeaves,spinach,sweetPotato"
res=requests.get(url, headers=header)
data=res.json()
print(data)
```

```python
vn=[]
veg=[]
d=data['columns']
dat=[]
mp=[]
rp=[]
smp=[]
u=[]
for i in data['columnMapping']:
    vn.append(i['vegName'])
l=len(vn)
l1=len(d)
for i in range(l): #loop on vegetable name
    for j in range(l1): #loop on each date
        rp.append(data["data"][i]['data'][j]['retailPrice'])
        veg.append(vn[i])
        dat.append(d[j])
        mp.append(data["data"][i]['data'][j]['y'])
        u.append(data["data"][i]['data'][j]['units'])
        smp.append(data["data"][i]['data'][j]['shopingMallPrice'])
        # print(vn[i],d[j],data["data"][i]['data'][j]['y'],data["data"][i]['data'][j]['units'],data["data"][i]['data'][j]['shopingMallPrice'],data["data"][i]['data'][j]['retailPrice'])
    print(len(rp),len(veg),len(dat),len(mp),len(u),len(smp))
```

```python
col={
    "Date":dat,
    "Vegetable_Name":veg,
    "Units":u,
    "Market_Price":mp,
    "Retail_Price":rp,
    "Shoping_Mall_Price":smp
}
dF=pd.DataFrame(col)
path="C:/Users/Madhavi/Downloads/Project/data_noida.csv"
dF.to_csv(path,index=False)
```

```python
Retail_Price_low_list=[]
Retail_Price_high_list=[]
for m in df_noida['Retail_Price'].str.split(' - '):
    for i in range(2):
        if i==0:
            Retail_Price_low_list.append(m[i])
        else:
            Retail_Price_high_list.append(m[i])
```

```python
df_noida['Retail_Price_low']=Retail_Price_low_list
df_noida['Retail_Price_high']=Retail_Price_high_list
```

```python
df_noida.info()
```

```python
df_noida['Retail_Price_low']=df_noida['Retail_Price_low'].astype(np.int64)
df_noida['Retail_Price_high']=df_noida['Retail_Price_high'].astype(np.int64)
```

```python
df_noida['Average_Retail_Price']=(df_noida['Retail_Price_low']+df_noida['Retail_Price_high'])/2
```

```python
df_noida.head()
```

```python
df_noida.drop(['Retail_Price'], axis=1,inplace=True)
```

```python
Shoping_Mall_Price_low_list=[]
Shoping_Mall_Price_high_list=[]
for m in df_noida['Shoping_Mall_Price'].str.split(' - '):
    for i in range(2):
        if i==0:
            Shoping_Mall_Price_low_list.append(m[i])
        else:
            Shoping_Mall_Price_high_list.append(m[i])
```

```python
df_noida['Shoping_Mall_Price_low']=Shoping_Mall_Price_low_list
df_noida['Shoping_Mall_Price_high']=Shoping_Mall_Price_high_list
```

```python
df_noida['Shoping_Mall_Price_low']=df_noida['Shoping_Mall_Price_low'].astype(np.int64)
df_noida['Shoping_Mall_Price_high']=df_noida['Shoping_Mall_Price_high'].astype(np.int64)
```

```python
df_noida['Average_Shoping_Mall_Price']=(df_noida['Shoping_Mall_Price_low']+df_noida['Shoping_Mall_Price_high'])/2
```
